*22.1 - ODEs*

A differential equation is a relationship between a function, f(x), its independent variable, x, and any number of its derivatives. An **ordinary differential equation** or ODE is a differential equation where the independent variable, and therefore also the derivatives, is in one dimension.

A **general solution** to a differential equation is a g(x) that satisfies the differential equation. Although there are usually many solutions to a differential equation, they are still hard to find. For an ODE of order n, a **particular solution** is a p(x) that satisfies the differential equation and n explicitly **known values** of the solution, or its derivatives, at certain points. A common set of known values for an ODE solution is the **initial value**. For an ODE of order n, the initial value is a known value for the 0th to (n−1)th derivatives at x=0,f(0),f(1)(0),f(2)(0),…,f(n−1)(0). Finding a solution to an ODE given an initial value is called the **initial value problem**.

*22.2 - Reduction of Order*

It is sometimes useful to **reduce the order** of higher order ODEs to first order ODEs. 
In order to do this we need to already know a solution. We then assume that a second solution will follow the form: y2(t)=y1(t)v(t). We then calculate the derivatives we will need and substitute the second assumed solution into the original ODE and solve for v(t) using a change of variable. After obtaining v(t), we can then use v(t) to expess a general or particular solution to the original ODE

*22.3 - Euler Method*

A commonly used method for solving ODEs is Euler's Method. The Euler Method uses linear approximation to approximate solutions to ODEs.

Starting from a given initial value of S0=S(t0), we can use the **Explicit Euler Formula** to integrate the states up to S(tf); these S(t) values are then an approximation for the solution of the differential equation. The Explicit Euler Formula is S(tj+1)=S(tj)+hF(tj,S(tj)). The Explicit Euler formula is the simplest and most intuitive method for solving initial value problems. At any state (tj,S(tj)) it uses F at that state to “point” toward the next state and then moves in that direction a distance of h.

To illustrate the method in python, we will consider the ODE df(t)/ft=sin(t) with a step size of 0.3



```
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
sns.set()

g = lambda t, s: np.sin(t) 
h1 = 0.3
t = np.arange(0, 1 + h1, h1) 
s0 = -1.0 


s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h1*g(t[i], s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'r--', label='Approximate')
plt.plot(t, -np.cos(-t), 'b', label='Exact')
plt.title('Approx and Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.show()
```

As we decrease the step size, the accuracy of the approximation increases. Here is the same ODE with a step size of 0.01. As we can see, this approximation is much more accurate.
