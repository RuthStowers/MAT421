**Chapter 17- Interpolation**

*17.1*

Assume data of independent data values xi and dependent data vlaues yi, where i = 1, ..., n. The goal is to estimate a function that goes through the data points. No underlying model is required, but it must still inform the user about the quality of the interpolation.

To begin we will use python to create an example of **interpolation**.

Example 1: Find the interpolation at x = 7.5 on the data x = [2, 3, 4] and y = [8, 7, 6] using python.

```python
import numpy as np
import matplotlib.pyplot as plt

x = [2, 3.5, 7]
y = [8, 7, 6]

x_new = 5.5
y_new = np.interp(x_new, x, y)
print(y_new)
```

```python
plt.plot(x, y, 'o-')
plt.title('Interpolation at x = 5.5')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

*7.2*

**Linear interpolation** is the estimated point that is assumed to lie on the line joining the nearest point to the left and right. 

Below we will show an example of linear interpolation. 

Example 2: Find the linear interpolation at x = 5.5 based on data x = [4,5,6], y = [5,7.5,7]. Create a graph.

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
```

```python
x = [4,5,6]
y = [5,7.5,7]

f = interp1d(x, y)
y_hat = f(5.5)
print(y_hat)
```

```python
plt.figure(figsize = (10, 8))
plt.plot(x, y, '-ob')
plt.plot(5.5, y_hat, 'ro')
plt.title('Linear Interpolation at x = 5.5')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

*7.3*

**Cubic spline interpolation** is the interpolating function as a set of piecewise cubic funtions. 

Example 3: Use the CubicSpline function to plot the cubic spline interpolation of the data set x = [5,5.5,7.5] and y = [2,4,3] for 4 <= x <= 8. 



```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [5,5.5,7.5]
y = [2,4,3]

f = CubicSpline(x, y, bc_type = 'natural')
x_new = np.linspace(4, 8, 100)
y_new = f(x_new)
```

```python
plt.figure(figsize = (10, 8))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
